```python
from sklearn.linear_model import LogisticRegression
import numpy as np

X = np.array([[1, 2], [2, 1], [3, 4], [4, 3], [7, 11], [8, 14]])
y = np.array([1, 1, 0, 0, 2, 2])
N=len(X)
ds=np.append(X,y.reshape(N,1),1)
print(ds)
```

```python
classes=list(set(y))
ds=X
for i in classes:
  yy=list()
  for j in y:
    if j==i:
      yy.append(1)
    else:
      yy.append(0)
  yy=np.array(yy)
  ds=np.append(ds,yy.reshape(N,1),1)
```

```python
ds
```

```python
models=list()
for i in range(3):
  lr_model=LogisticRegression()
  lr_model.fit(ds[:,0:2],ds[:,2+i])
  models.append(lr_model)
```

```python
def prob(z):
  totl=np.sum(np.exp(z))
  y_pred=list()
  for i in z:
    y_pred.append(np.exp(i)/totl)
  #y_pred=np.array(y_pred)
  return y_pred
  
```

```python
def predict(X,models):
  z=list()
  for i in range(3):
    z.append(np.sum(models[i].coef_*X)+models[i].intercept_)
  z=np.array(z)

  y_pred=prob(z)

  return y_pred.index(max(y_pred))
```

```python
X=np.array([8,16])
predict(X,models)
```
